```python
%config IPCompleter.greedy=True
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset=pd.read_excel('C:/Users/mert_/Desktop/uygulama_dersleri/Veriler/kredi_veriseti.xlsx')
dataset.head()
```

```python
one_hot= OneHotEncoder(handle_unknown='ignore')
```

```python
one_hot_cinsiyet=one_hot.fit_transform(dataset['cinsiyet'].values.reshape(-1,1)).toarray()
one_hot_cinsiyet
```

```python
one_hot.categories_
```

```python
one_hot_df=pd.DataFrame(one_hot_cinsiyet,columns=one_hot.categories_)
one_hot_df.head()
```

```python
one_hot_df2=dataset.join(one_hot_df)
one_hot_df2.head()
```

```python
one_hot_df2.drop('cinsiyet',axis=1,inplace=True)
one_hot_df2.head()
```

```python
X=one_hot_df2.drop('kredi',axis=1)
y=one_hot_df2['kredi']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
model=LogisticRegression(random_state=0)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
